```python
import numpy as np
x = np.array([['Sunny', 'Warm', 'Normal', 'Strong', 'Warm', 'Same', 'yes'],
             ['Sunny', 'Warm', 'High', 'Strong', 'Warm', 'Same', 'yes'],
             ['Rainy', 'Cold', 'High', 'Strong', 'Warm', 'Change', 'no'],
             ['Sunny', 'Warm', 'High', 'Strong', 'Cool', 'Change', 'yes']])
#print(x)
#Calculating number of rows and columns
r, c = x.shape
print("The number of rows and columns are: ", r, "and", c-1)
print("The dataset is: \n",x)

s = np.empty(c-1, dtype=object)
for i in range(c-1): 
    s[i]="\u03A6"
print("Most specific hypothesis \n",s)

g = np.empty(c-1, dtype=object)
for i in range(c-1): 
    g[i]="?"
print("Most specific hypothesis \n",s)

for i in range(c-1):
    s[i] = x[0,i]
print("After processing first training sample \n",s)

flag=0
for i in range(1,r):
    if x[i,c-1]=='yes':
        for j in range(c-1):
            if x[i,j] != s[j]:
                s[j] = '?'
        print("Intermediate specific hyothesis : \n", s)
        if flag==0:
            flag=1
            continue
        else:
            r1=np.shape(g)[0]
            for i1 in range(r1):
                for j1 in range(c-1):
                    if g[i1,j1]!=x[i1,j1] and g[i1,j1]!="?":
                        g = np.delete(g,i1,axis=0)
    else:
        for j in range(c-1):
            g1=np.empty(c-1, dtype=object)
            if s[j]=='?':
                continue
            elif x[i,j]!=s[j]:
                for l in range(c-1):
                    g1[l] = "?"
                g1[j]=s[j]
                g = np.vstack([g, g1])
        g=np.delete(g,(0), axis=0)
        print("\nIntermediate general hypothesis\n", g)
        
print("Final Specific Hypothesis set: \n", s)
print("Final General Hypothesis set: \n", g)

row_g, col_g = g.shape
print("\nEnter new instance value")
new = []
new = np.empty(c-1, dtype = object)
for i in range(c-1):
    new[i] = input("Enter attribute value : ")
print("\nThe given instance is ", new)
flag = 1
for i in range(c-1):
    if s[i] != '?' and s[i] != new[i]:
        flag = 0
        break
flagg = 1
for i in range(row_g):
    for j in range(col_g-1):
        if g[i][j]!=new[j] and g[i][j]!=0:
            flagg = 0
            break
    if flagg==1:
        break
print("\nThe assigned lable for ENJOYING WATER SPORT is \n")
if flag==1 or flagg==1:
    print("YES")
else:
    print("NO")
```
